#                       Regressions

```python
# Importing the library need for sampling the train and test data
from sklearn.model_selection import train_test_split
import math

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 42)
```

MAE: measure the error of our predicting model

MSE: is a good metric to evaluate the accuracy of the model

R^2: is a measure of how close each data point fits to the regression line (0-1)

RMSE: measures the average difference between values predicted by a model and the actual values.

## Linear Regression

```python
# Importing the LinearReg and MSE, MAE, R^2 libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import seaborn as sns
```

```python
# Creating Linear Regression function
def LR(X_train, X_test, y_train, y_test):
  linReg = LinearRegression()
  linReg.fit(X_train, y_train)              # Fit linear model

  y_pred = linReg.predict(X_test)           # Predict using the linear model

  pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
  print(pred_df)
  print("MAE: ", round(mean_absolute_error(y_test, y_pred), 3))
  print("MSE: ", round(mean_squared_error(y_test, y_pred), 3))
  print("R^2: ", round(r2_score(y_test, y_pred), 3))
  print('RMSE: ', round(math.sqrt(mean_squared_error(y_test, y_pred)), 3))

  sns.regplot(x=y_test,y=y_pred,ci=None,color ='red')
  plt.title("Linear Regression")
  plt.xlabel("Actual")
  plt.ylabel("Predicted")
```

```python
# Apply the LR on our data
LR(X_train, X_test, y_train, y_test)
```

## Support Vector Regression

```python
# Importing the SVR library
from sklearn.svm import SVR
```

```python
# Creating Support Vector Regression function
def support_reg(X_train, X_test, y_train, y_test):
  svr = SVR(C=1.0, epsilon=0.2)         # Optimizaton values
  svr.fit(X_train, y_train)

  y_pred = svr.predict(X_test)

  pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
  print(pred_df)
  print("MAE: ", round(mean_absolute_error(y_test, y_pred), 3))
  print("MSE: ", round(mean_squared_error(y_test, y_pred), 3))
  print("R^2: ", round(r2_score(y_test, y_pred), 3))
  print('RMSE: ', round(math.sqrt(mean_squared_error(y_test, y_pred)), 3))

  sns.regplot(x=y_test,y=y_pred,ci=None,color ='blue')
  plt.title("Support Vector Regression")
  plt.xlabel("Actual")
  plt.ylabel("Predicted")
```

```python
# Apply the Support Vector Reg on our data
support_reg(X_train, X_test, y_train, y_test)
```

## Lasso Regression

```python
# Importing the lasso library
from sklearn import linear_model
```

```python
# Creating Lasso Regression function
def lasso(X_train, X_test, y_train, y_test):
  clf = linear_model.Lasso(alpha=0.1)
  clf.fit(X_train, y_train)

  y_pred = clf.predict(X_test)

  pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
  print(pred_df)
  print("MAE: ", round(mean_absolute_error(y_test, y_pred), 3))
  print("MSE: ", round(mean_squared_error(y_test, y_pred), 3))
  print("R^2: ", round(r2_score(y_test, y_pred), 3))
  print('RMSE: ', round(math.sqrt(mean_squared_error(y_test, y_pred)), 3))

  sns.regplot(x=y_test,y=y_pred,ci=None,color ='green')
  plt.title("Lasso Regression")
  plt.xlabel("Actual")
  plt.ylabel("Predicted")
```

```python
# Apply the Lasso Reg on our data
lasso(X_train, X_test, y_train, y_test)
```

# Optimization

## Stochastic Gradient Descent (SGD)

It is used to minimize the difference between y and y_hat.

```python
from sklearn.linear_model import SGDRegressor

# Creating SGD function
def SGD(X_train, X_test, y_train, y_test):
  reg = SGDRegressor(learning_rate='adaptive', penalty='l2')
  #SGDRegressor(max_iter=1000, tol=1e-3)
  reg.fit(X_train, y_train)

  y_pred = reg.predict(X_test)

  pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
  print(pred_df)
  print("MAE: ", round(mean_absolute_error(y_test, y_pred), 3))
  print("MSE: ", round(mean_squared_error(y_test, y_pred), 3))
  print("R^2: ", round(r2_score(y_test, y_pred), 3))
  print('RMSE: ', round(math.sqrt(mean_squared_error(y_test, y_pred)), 3))

  sns.regplot(x=y_test,y=y_pred,ci=None,color ='brown')
  plt.title("Stochastic Gradient Descent")
  plt.xlabel("Actual")
  plt.ylabel("Predicted")
```

```python
# Original data
SGD(X_train, X_test, y_train, y_test)
```

```python
# Original data and Standard scaler
SGD(X_train_std, X_test_std, y_train, y_test)
```

```python
# Original data and MinMax scaler
SGD(X_train_mm, X_test_mm, y_train, y_test)
```

```python
# New Features
SGD(X_train_feat, X_test_feat, y_train, y_test)
```

```python
# New Features and Standard scaler
SGD(X_train_std_feat, X_test_std_feat, y_train, y_test)
```

```python
# New Features and MinMax scaler
SGD(X_train_mm_feat, X_test_mm_feat, y_train, y_test)
```

We notice that our predicted values are very large with using normal standardiziation. However, the values dropped significatly when features are scaled. It seems that SGD is sensitive to the features with outliers.
